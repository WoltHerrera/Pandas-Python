# 03 - Limpieza con dropna y drop_duplicates

Antes de analizar datos conviene revisar nulos, filas vacias y duplicados.

```python
import pandas as pd

jugadores = pd.DataFrame({
    'jugador': ['Trauco', 'Advincula', None, 'Lapadula', 'Trauco', None],
    'posicion': ['Defensa', 'Defensa', None, 'Delantero', 'Defensa', None],
    'pais': ['Peru', 'Peru', None, 'Peru', 'Peru', None],
})
jugadores
```

## Diagnosticar nulos

```python
jugadores.isna().sum()
```

## Eliminar filas completamente vacias

```python
sin_filas_vacias = jugadores.dropna(how='all')
sin_filas_vacias
```

## Eliminar filas con cualquier nulo

```python
jugadores.dropna()
```

## Eliminar duplicados

```python
sin_duplicados = sin_filas_vacias.drop_duplicates()
sin_duplicados
```

## Eliminar columnas

```python
sin_duplicados.drop(['posicion', 'pais'], axis=1)
```
