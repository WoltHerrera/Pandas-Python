# 02 - Concat y union

`pd.concat()` permite apilar tablas con la misma estructura. Es util para juntar datos provenientes de diferentes archivos, meses o paises.

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path('..') / 'data'
jugadores_peru = pd.read_csv(DATA_DIR / 'jugadores_a.csv')
jugadores_brasil = pd.read_csv(DATA_DIR / 'jugadores_b.csv')
```

```python
jugadores_peru
```

```python
jugadores_brasil
```

## Union all

Concatena todas las filas y conserva los indices originales.

```python
pd.concat([jugadores_peru, jugadores_brasil])
```

## Reindexar

`ignore_index=True` crea un indice nuevo y ordenado.

```python
union_all = pd.concat([jugadores_peru, jugadores_brasil], ignore_index=True)
union_all
```

## Union sin duplicados

Pandas no tiene una funcion llamada `union` para dataframes, pero puedes simularla con `concat()` y `drop_duplicates()`.

```python
union_all.drop_duplicates()
```
