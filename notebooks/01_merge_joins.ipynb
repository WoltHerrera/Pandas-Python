# 01 - Merge y joins

En este notebook aprenderas a combinar tablas relacionadas con `pd.merge()`. El ejemplo usa clientes y pedidos para mostrar que ocurre con `inner`, `left`, `right` y `outer`.

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path('..') / 'data'
clientes = pd.read_csv(DATA_DIR / 'clientes.csv')
pedidos = pd.read_csv(DATA_DIR / 'pedidos.csv')
```

```python
clientes
```

```python
pedidos
```

## Inner join

Devuelve solo los registros que tienen coincidencia en ambas tablas.

```python
pd.merge(clientes, pedidos, how='inner', on='cliente_id', indicator=True)
```

## Left join

Mantiene todos los clientes. Si un cliente no tiene pedido, las columnas del pedido quedan vacias.

```python
pd.merge(clientes, pedidos, how='left', on='cliente_id', indicator=True)
```

## Right join

Mantiene todos los pedidos, incluso si el cliente no existe en la tabla de clientes.

```python
pd.merge(clientes, pedidos, how='right', on='cliente_id', indicator=True)
```

## Outer join

Mantiene todos los registros de ambas tablas.

```python
pd.merge(clientes, pedidos, how='outer', on='cliente_id', indicator=True)
```

## Columnas con distinto nombre

Cuando la clave no tiene el mismo nombre en ambas tablas, usa `left_on` y `right_on`.

```python
clientes_con_codigo = clientes.rename(columns={'cliente_id': 'codigo_cliente'})
pd.merge(
    clientes_con_codigo,
    pedidos,
    how='left',
    left_on='codigo_cliente',
    right_on='cliente_id',
    indicator=True,
)
```
