# 05 - Sample

`sample()` permite tomar una muestra aleatoria de filas. Usaremos `random_state` para que el resultado sea reproducible.

```python
from pathlib import Path
import numpy as np
import pandas as pd

DATA_DIR = Path('..') / 'data'
ventas = pd.read_csv(DATA_DIR / 'ventas.csv')
ventas
```

## Muestra por porcentaje

```python
ventas.sample(frac=0.25, random_state=42)
```

## Muestra por cantidad fija

```python
ventas.sample(n=3, random_state=42)
```

## Numeros aleatorios con NumPy

Se usa `numpy` directamente. Evitamos `pd.np` porque esta obsoleto.

```python
numeros = pd.DataFrame({'valor': np.random.default_rng(seed=42).random(10)})
numeros
```
