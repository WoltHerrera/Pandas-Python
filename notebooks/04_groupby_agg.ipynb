# 04 - GroupBy y agregaciones

`groupby()` permite resumir datos por una o varias columnas. En este ejemplo calcularemos ventas por categoria y pais.

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path('..') / 'data'
ventas = pd.read_csv(DATA_DIR / 'ventas.csv')
ventas['importe'] = ventas['cantidad'] * ventas['precio_unitario']
ventas
```

## Conteo por categoria

```python
ventas.groupby('categoria').count()
```

## Resumen con agg

La sintaxis con nombres permite que el resultado sea mas claro.

```python
ventas.groupby('categoria').agg(
    total_vendido=('importe', 'sum'),
    ticket_promedio=('importe', 'mean'),
    cantidad_operaciones=('venta_id', 'count'),
    productos=('producto', lambda valores: ', '.join(sorted(valores.unique()))),
)
```

## Agrupar por mas de una columna

```python
ventas.groupby(['pais', 'categoria']).agg(
    total_vendido=('importe', 'sum'),
    unidades=('cantidad', 'sum'),
)
```
